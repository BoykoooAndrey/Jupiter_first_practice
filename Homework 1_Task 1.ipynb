Определить промежутки, на котором f < 0

```python
from sympy import *

a = 0.6
b = 5.5
c = 10
d = -5

x = Symbol('x')

f_x = a * (x**3) + b * (x**2) + c * (x**1) - d
```

```python
f_x
```

```python
print('Корень №1:', solve(f_x,x)[0].n())
print('Корень №1:', solve(f_x,x)[1].n())
```

```python
from sympy.plotting import plot
g1 = plot(f_x,(x,-100,100),line_color = 'red', title = 'График функции')
```

```python
derivative = diff(f_x)
derivative
```

```python
g2 = plot(derivative,(x,-100,100),line_color = 'red', title = 'График функции')
```

```python
g3 = plot(f_x, derivative,(x,-100,100), title = 'Совмещенный график', legend = True, 
           xlable ='X', ylable = 'Y', show = False)
g3[0].line_color = 'red'
g3[1].line_color = 'black'
g3.show()
```

```python
solve(derivative > 0,x)
```

```python
solve(derivative < 0,x)
```
